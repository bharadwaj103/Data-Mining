```python
# prompt: take a data of a starup and analyze its groth highs and lows.Profits and losses year wise. investment  show the output in the bar graph,grpahs without any errors in the output.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample startup data (replace with your actual data)
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Revenue': [100000, 150000, 200000, 250000, 300000],
    'Expenses': [80000, 120000, 180000, 220000, 240000],
    'Investment': [50000, 75000, 100000, 125000, 150000],
    'Profit': [20000, 30000, 20000, 30000,60000]
}

df = pd.DataFrame(data)

# Calculate profit/loss
# df['Profit/Loss'] = df['Revenue'] - df['Expenses']


# Create subplots for better visualization
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 10))

# Plot revenue and expenses over time
axes[0, 0].plot(df['Year'], df['Revenue'], label='Revenue', marker='o')
axes[0, 0].plot(df['Year'], df['Expenses'], label='Expenses', marker='o')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Amount')
axes[0, 0].set_title('Revenue vs. Expenses')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Bar chart for profit/loss
axes[0, 1].bar(df['Year'], df['Profit'], label='Profit', color='green')
# axes[0, 1].bar(df['Year'], df[df['Profit/Loss'] < 0]['Profit/Loss'], label='Loss', color='red') #Plotting losses
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Amount')
axes[0, 1].set_title('Profit')
axes[0, 1].legend()
axes[0,1].grid(True)



# Investment trend
axes[1, 0].plot(df['Year'], df['Investment'], marker='o', color='purple')
axes[1, 0].set_xlabel("Year")
axes[1, 0].set_ylabel("Investment")
axes[1,0].set_title("Investment Trend")
axes[1, 0].grid(True)

# Remove the unused subplot
axes[1, 1].axis('off')


plt.tight_layout() # Adjusts spacing between subplots
plt.show()
```
